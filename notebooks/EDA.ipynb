Exploratory Data Analysis – Telco Customer Churn


```python
#importing relevant libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

loading the dataset

```python
df = pd.read_csv("../data/WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head()
```

checking if there are any missing values

```python
df.isna().sum()
```

Structure and summery 

```python
df.shape
df.info()
df.describe(include="all")
```

Converting non numerical values to Na

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

df['TotalCharges'].isna().sum()
```

checking if there are any duplicate values

```python
df.duplicated().sum()
```

Fixing TotalCharges

```python
df = df.dropna(subset=['TotalCharges'])
df.shape
```

Noticed class imbalance in Churn

```python
df['Churn'].value_counts(), df['Churn'].value_counts(normalize=True)
```

Comparing yes and no
#Churn Distribution Plot

```python
plt.figure(figsize=(5,4))
sns.countplot(data=df, x='Churn')
plt.title("Churn Distribution")
plt.show()
```

```python
num_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']

df[num_cols].hist(bins=25, figsize=(12,6))
plt.suptitle("Numerical Feature Distributions")
plt.show()
```

Numerical Features vs Churn

```python
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(data=df, x='Churn', y=col)
    plt.title(f"{col} vs Churn")
    plt.show()
```

```python
cat_cols = [c for c in df.columns 
            if df[c].dtype == 'object' and c not in ['customerID', 'Churn']]
cat_cols
```

Categorical Features vs Churn

```python
for col in cat_cols:
    plt.figure(figsize=(7,4))
    sns.countplot(data=df, x=col, hue='Churn')
    plt.title(f"Churn by {col}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[num_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation between Numerical Features")
plt.show()
```
